```python
print('Matches played so far:',match_data.shape[0])
print('\n Cities played at:',match_data['city'].unique())
print('\n Teams participated:',match_data['team1'].unique())
```

```python
match_data['Season']=pd.DatetimeIndex(match_data['date']).year
match_data.head()
```

```python
match_per_season=match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
sns.countplot(match_data['Season'])
plt.xticks(rotation=45,fontsize=10)
plt.xlabel('Season',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Total matches played in each Season',fontsize=10,fontweight="bold")
```

```python
season_data=match_data[['id','Season']].merge(ball_data,left_on='id',right_on='id',how='left').drop('id',axis=1)
season_data.head()
```

```python
season=season_data.groupby(['Season'])['total_runs'].sum().reset_index()
p=season.set_index('Season')
ax=plt.axes()
ax.set(facecolor="grey")
sns.lineplot(data=p,palette='magma')
plt.title('Total runs in each season',fontsize=12,fontweight="bold")
plt.show()
```

```python
runs_per_season=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs_per_season['Runs scored per match']=runs_per_season['total_runs']/runs_per_season['matches']
runs_per_season.set_index('Season',inplace=True)
runs_per_season
```

```python
toss=match_data['toss_winner'].value_counts()
ax=plt.axes()
ax.set(facecolor='grey')
sns.set(rc={'figure.figsize':(10,5)},style='darkgrid')
ax.set_title("No of tosses won by each team",fontsize=15,fontweight="bold")
sns.barplot(y=toss.index,x=toss,orient='h',palette='icefire')
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()
```

```python
ax=plt.axes()
ax.set(facecolor="grey")
sns.countplot(x="Season",hue="toss_decision",data=match_data,palette='magma')
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n Season',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision across seasons',fontsize=12,fontweight="bold")
plt.show()
```

```python
sns.countplot(match_data['winner'][match_data['eliminator']=='Y'],data=match_data)
plt.title('Most wins in Eliminator',fontsize=1)
plt.xticks(rotation=90)
plt.show()
```

```python
teams=match_data['toss_winner'].unique()
decision_making=pd.DataFrame([],columns=['Toss Winner','Decision','Times'])
for id,element in enumerate(teams):
  temp_bat=match_data[(match_data['toss_winner']==element) & (match_data['toss_decision']=='bat')]
  temp_fiels=match_data[(match_data['toss_winner']==element) & (match_data['toss_decision']=='field')]
  decision_making=decision_making.append({'Toss Winner':element,'Decision':'bat','Times':temp_bat['toss_winner'].count()},ignore_index=True)
  decision_making=decision_making.append({'Toss Winner':element,'Decision':'field','Times':temp_fiels['toss_winner'].count()},ignore_index=True)
```

```python
print(decision_making)
```

```python
sns.catplot(x="Toss Winner",y="Times",hue="Decision",data=decision_making,kind="bar",height=5,aspect=2)
plt.xticks(rotation=90)
plt.title("Toss decision of Teams")
plt.xlabel("IPL Teams")
plt.ylabel("Decision")
```
